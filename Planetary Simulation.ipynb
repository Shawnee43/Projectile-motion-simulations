```python
import numpy as np
import matplotlib.pyplot as plt

#Parametres 
G = 6.67e-11 
M = 1.989e30
dt = 3600
steps = 10000


#Initial Poition and Velocity (Earth)
r = np.array([1.5e11, 0])
v = np.array([0,3e4])

#Store Positions 
x_vals = []
y_vals = []

#Position and velocity loop
for _ in range(steps):
    distance = np.linalg.norm(r)
    accel = -G * M * r / distance**3

    
#Update position and velocity
    v += accel * dt
    r += v * dt

#Save positions
    x_vals.append(r[0])
    y_vals.append(r[1])

plt.figure(figsize=(6,6))
plt.plot(x_vals, y_vals, label='Planet Path')
plt.plot(0, 0, 'yo', label='Sun')  # Sun at origin
plt.xlim(-2e11, 2e11)
plt.ylim(-2e11, 2e11)
plt.axis('equal')
plt.grid(True)
plt.xlabel("x position (m)")
plt.ylabel("y position (m)")
plt.title("Gravitational Orbit Simulation")
plt.legend()
plt.show()



```
