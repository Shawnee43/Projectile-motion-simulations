```python
import numpy as np
import matplotlib.pyplot as plt

g= 9.81 #gravity 
a=50 #acceleration 
v0=0 #velocity in m/s
Theta= np.radians(45) #launch angle 

t= np.arange(0,120,10)
vel=v0+a*t
x= vel * np.cos(Theta)
y= vel * np.sin(Theta)*t - 0.5*g*t**2

plt.plot(x, y)
plt.title("Projectile Motion")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.grid()
plt.show()
```
