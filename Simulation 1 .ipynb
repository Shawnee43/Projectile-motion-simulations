```python
import numpy as np
import matplotlib.pyplot as plt

g= 9.81 #gravity 
v= 50 #velocity in m/s
Theta= np.radians(45) #launch angle 

t= np.linspace(0,10,100)
x= v * np.cos(Theta) * t
y= v * np.sin(Theta) * t * g * t**2

plt.plot(x, y)
plt.title("Projectile Motion")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.grid()
plt.show()
```
