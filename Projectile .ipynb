```python
import numpy as np
import matplotlib.pyplot as plt

g= 9.81 #gravity 
v0=69.2964 #velocity in m/s
Theta= np.radians(45) #launch angle 

t= np.arange(0,10,0.1)
x= v0 * np.cos(Theta)*t
y= v0 * np.sin(Theta)*t - 0.5*g*t**2

plt.plot(x, y)
plt.title("Projectile Motion")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.grid()
plt.show()

```
