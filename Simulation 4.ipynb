```python
import numpy as np 
import matplotlib.pyplot as plt 

#Parametres
R=1
w=2
t=np.arange(0,5,0.1)

#Position vectors
x = R*np.cos(w*t)
y = R*np.sin(w*t)

#Velocity vectors 
dx = -w*R*np.sin(w*t)
dy = w*R*np.cos(w*t)

#Acceleration vectors
d2x = -w**2*R*np.cos(w*t)
d2y = -w**2*R*np.sin(w*t)

plt.plot(x,y)
plt.quiver(x, y, dx, dy, color='blue', scale=20)
plt.quiver (x, y, d2x, d2y, color='orange', scale=20)
plt.title('Uniform Circular Motion')
plt.grid(True)
```
